```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('pima-indians-diabetes.csv')
df.head()
```

```python
df.columns = ['pregnancies', 'glucose', 'blood_pressure', 'skin_thickness', 'insulin', 
                'bmi', 'diabetes_pedigree', 'age_yr', 'outcome']
df.head()
```

```python
def central_tendency(col):
    mean = df.describe()[col]['mean']
    median = df.describe()[col]['50%']
    std = df.describe()[col]['std']
    min_ = df.describe()[col]['min']
    max_ = df.describe()[col]['max']
    
    fig, ax = plt.subplots(1,1, figsize=(16,7))
    sns.distplot(df[col], bins = 20)
    ax.axvline(mean, color = 'r', linestyle = '--', label = 'mean')
    ax.axvline(median, color = 'g', linestyle = '--', label = 'median')
    ax.axvline(min_, color = 'b', linestyle = '-.', label = 'min')
    ax.axvline(max_, color = 'r', linestyle = ':', label = 'max')
    ax.legend()
    plt.show()
    
    print(f'Mean of {col} is {mean}')
    print(f'Median of {col} is {median}')
    print(f'Min of {col} is {min_}')
    print(f'Max of {col} is {max_}')
    print(f'Std of {col} is {std}')
    
    plt.figure(figsize = (8,6))
    sns.boxplot(df[col])
    plt.show();

    
central_tendency('pregnancies')
```
